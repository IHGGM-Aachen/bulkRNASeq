} else if(condition2) {

#### Exersice:  Write a function to compotes the nth digit of the Fibonacci sequence such that the first number in the sequence is 0, the second number is 1, and then all proceeding numbers are the sum of the n - 1 and the n - 2 Fibonacci number.

* The Fibonacci sequence is a sequence of integers that starts: 1, 2, 3, 5, 8 where each proceeding integer is the sum of the previous two integers

```r
# Generate your code here
```

```r
fib <- function(n){
    stopifnot(n > 0)
    if(n == 1){
        1
    } else if(n == 2){
        2
    } else {
        fib(n - 1) + fib(n - 2) #self-calling function does not need "return"
    }
}

print(fib(1))
print(fib(2))
print(fib(3))
print(fib(4))
print(fib(5))
print(fib(6))
print(fib(7))
```

#### Exersice:  write a funxction to calculate the degrees of a graph. Suppose only the adjacency matrix is available.

* Use graph.adjacency function

```r
# Generate your code here
```

```r
calculate_degrees <- function(adj_matrix) {
    graph <- graph.adjacency(adj_matrix, mode = "undirected", weighted = NULL)#Create a graph from the adjacency matrix 
    degrees <- degree(graph) #Calculate degrees  
    return(degrees)
}

adjacency_matrix <- matrix(c(0, 1, 1, 1, 0, 0, 1, 0, 0), nrow = 3, byrow = TRUE) #Example
degrees <- calculate_degrees(adjacency_matrix) # Call the function
print(degrees) #Print the result
```

#### Exersice: write a function to calculate the degrees of a graph. Suppose only the adjacency matrix is available. while you work with genomic data, the dataset contains some thousand genes, therefore, graph.adjacency function can not handle it. Therefore you might write a code to calculate the degrees.

```r
# Generate your code here
```

```r
degrees_per_node = function(mat){
    d <- dim(mat)[1] # number of vertices
    degrees = matrix(0, nrow = d, ncol = 1) 
    for (i in 1:nrow(mat)){ 
        for (j in 1:ncol(mat)){ 
            if (i == j){ 
                degrees[i] = degrees[i] + 2 * mat[i, j] #This deals with loops 
            }else{ 
                degrees[i] = degrees[i] + mat[i, j] 
            } 
        } 
    } 
    return(degrees)
} 

set.seed(123) #The set.seed() function in R is used to create reproducible results
random_edges <- sample(0:1, 16, replace = TRUE) #generate a random directed graph
mat <- matrix(random_edges, nrow = 4) #generate adjacency matix of the graph

print(mat)
print(degrees_per_node(mat))
```

#### Exersice: Write a function to determine the second neighborhood a given node  in an undirected graph. 
* Only the adjacencymatrix is available
* Do you know any biological examples? Why do we look for the i-th neighbourhood of a node?

```r
# Generate your code here
```

```r
second_neighbors <- function(adj,v){
    lst <- c() 
    for (i in 1:nrow(adj)){
        for (j in 1:ncol(adj)){ 
            if (adj[v, i] * adj[i, j] > 0){ 
                lst = c(lst, j)
            } 
        } 
    } 
    return(unique(lst)) 
}

tr <- make_tree(13, children = 3, mode = "undirected") #generate an example
adj_mat <- get.adjacency(tr) 
second_neighbors(adj_mat,3)
plot(tr)
```

#### Exersice: Write a function that takes a graph  and a pair of nodes  and . Determine the length of shortest paths between the nodes u and v.  
* Only the adjaency matrix is available and is not possilbe to use shortest.paths function

```r
shortest_path <- function(mat){ 
    d <- dim(mat)[1] # number of vertices
    sh_path = matrix(Inf, nrow = d, ncol = d) # make an empty matrix to store hortest pathes
    for(i in 1:d){
        sh_path[i,i] = 0        
    } 
    
    nw <- mat #initialization
    for(i in 1:d){ #repeat it for the number of nodes times
        for(j in 1:d){# repeat it for rows of the matrix mat
            for(k in 1:d){ # repead it for the columns of matrix mat
               if(sh_path[j,k] == Inf & nw[j,k] == 1){ # fined new shortest paths between two nodes j and k
                    sh_path[j,k] <- i # the length of the shortests path
                }
            }
        }
        nw <- nw %*% mat
        
    }    
    return(sh_path)
}

tr <- make_tree(13, children = 3, mode = "undirected") #generate an example
mat <- get.adjacency(tr)
plot(tr)
print(shortest_path(mat))
```

#### Exersice:  Write a function to determine the most similar pair of nodes, i.e. nodes which share the largest number of neighbors relative to the number of nodes in the union of neighbors in a given undirected graph. 
* use neighbors function

```r
# Generate your code here
```

```r
the_most_similar_pair_of_nodes <- function(g){ 
    adj <- get.adjacency(g) #adjacency matrix
    lst = matrix(0, 0, 2) # makes a list if nodes with the most similrity
    max_sim = 0 # maximum similarity value
    for (i in 1:(nrow(adj) - 1)){  
        ni = neighbors(g, i) # neighbors of node i
        for (j in (i + 1):nrow(adj)){ 
            nj = neighbors(g, j) # neighbors of node j
            uni <- union(ni, nj) # union of neighbors of nodes i and j 
            int <- intersect(ni, nj) # mutual neighbors of nodes i and j 
            if (length(uni) > 0 & length(int) > 0){ 
                sim = length(int) / length(uni) # similarity measure for nodes i and j
                if (sim == max_sim){ # if the similarity is maximum then  the pair i and j added to the list
                    lst <- rbind(lst, c(i, j))
                }
                if (sim > max_sim){ # if the similarity is more then known maximum then the list is renewed
                    lst = c(i, j) # new list is built
                    max_sim = sim # new maximum similarity value stored
                } 
            } 
        } 
    } 
    return(lst) 
} 

tr <- make_tree(13, children = 3, mode = "undirected") #generate an example
plot(tr)
print(the_most_similar_pair_of_nodes(tr))
```
